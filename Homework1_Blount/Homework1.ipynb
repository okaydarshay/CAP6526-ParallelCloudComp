```python
from pyspark.sql import SparkSession
from pyspark.conf import SparkConf
```

```python
#remove punctuation
def removePunctuation(text):
    punc = '!"#$%&\()*+,./:;<=>?@[\\]^_~`{|}='
    lowercase = text.lower()
    for ch in punc:
        lowercase = lowercase.replace(ch,'')
    return lowercase
```

```python
hamlet_lc = hamlet.map(removePunctuation)
```

```python
hamlet_lc.collect()
```

```python
import itertools
```

```python
word_count = (hamlet_lc.filter(lambda x: x != "")
                    .map(lambda x: x.split(" "))
                    .flatMap(lambda x: itertools.combinations(x,2))
                    .filter(lambda x: x[0] != "")
                    .map(lambda x: (x,1))
                    .reduceByKey(lambda x, y: x + y).collect()
             )
```
